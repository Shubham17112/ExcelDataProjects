```python
'''
data has null values so remove it
''' 

df.dropna(inplace=True)
df.info()
```

```python
'''
column 3: overview don't need 

column 8: poster url don't nee
'''
df.drop(['Poster_Url','Overview'],inplace=True,axis=1)
```

```python
df
```

```python
'''
column 1: we have release data  yyy mm dd ( object ) covert into date
column 2: we have release data  title (object ) covert into string

column 4: poluarity (object) covert into int
column 5: vote  (object) covert into int
column 6:  lang object -> str
column 7: genre we have to break it (object ) -> str

'''

df['Release_Date'] = pd.to_datetime(df['Release_Date'],errors='coerce')
df['Title'] = df['Title'].astype('str')
df['Popularity'] = df['Popularity'].astype('int')
df['Vote_Count'] = df['Vote_Count'].astype('int')
df['Vote_Average'] = df['Vote_Average'].astype('float')
df['Original_Language'] = df['Original_Language'].astype('str')
df['Genre'] = df['Genre'].astype('str')



df.info()
```

```python
df
```

```python
df['Genre'] = df['Genre'].apply( lambda x: x.split(', ') ) 
df
```

```python
'''  column 7: genre we have to break it  '''

 
# df['Genre'].explode()  thi will explode only gnere column and u will lost rest 
df = df.explode('Genre').reset_index(drop=True)

df
```

```python
df.info()
```

```python
df['Vote_Average'].describe()
```

```python
'''
we will categorize vote avg  into 3 catagories  popular , avagare and below avarege

'''

def categorize(col):
    labels = ['Below Avg', 'Avg', 'Above Avg', 'High']
    edges = [
        0,  # start from 0
        df[col].describe()['25%'],
        df[col].describe()['50%'],
        df[col].describe()['75%'],
        df[col].max()  # end at max value
    ]
    df['Vote_Average_Level'] = pd.cut(df[col], bins=edges, labels=labels, include_lowest=True)
    return df['Vote_Average_Level']
```

```python
categorize('Vote_Average')
df
```
