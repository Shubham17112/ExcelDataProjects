```python
# Line Plot 
# genrally it's used to show trend over a time  or comparing values
import seaborn as sns
import matplotlib.pyplot as plt

x = [1,2,3,4,5]

y =  [ 2,34,4,5,6]


sns.lineplot(x=x,y=y)
plt.show()
```

```python
# giving labesl only possible with metplotlib
import matplotlib.pyplot as plt
import seaborn as sns
x = [1,2,3,4,5]

y =  [ 2,34,4,5,6]


sns.lineplot(x=x,y=y)

plt.xlabel("X Axis Name")   #  Matplotlib
plt.ylabel("Y Axis Name")   #  Matplotlib
plt.title("Line Plot Example")
plt.show()
```

```python
#  Add Multiple Lines (Comparing Categories)


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.DataFrame({
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri"] * 2,
    "sales": [120, 135, 150, 170, 160, 100, 115, 140, 130, 150],
    "store": ["A"]*5 + ["B"]*5
})

# sns.lineplot(data = df,y= "sales",x = "day",hue="store")


# styling 
# maker,style ,palette

# instead of line plot use replot
sns.relplot(data = df,y= "sales",x = "day",hue="store",marker=True,kind="line")
```

```python
# Histogramme plot.
# it's allow to show u range dat like u have 100 student score data so u can plot like this 0-10 socre studetns , 10-20 score students and so on.


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = sns.load_dataset('iris')

# print(data.to_string())

sns.histplot(data = data, x = 'sepal_length',hue='species',bins=5,kde=True)

# here bins helps u to anaylis petal leght ex: let's suppose u have 1 cm - 7.5 cm petal leths now u divded into 5 bins then 
# each bins will cary 7.5/5 cm petal leghts and the flower fall under this bins will count 
# elemente = 'step' will fade the line between bars

# kde will draw that somoth curv which tell the count of spetal letnh across each bin 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.randn(1000)

# Histogram with element='step'
sns.histplot(data=data, bins=20, element='step')

# Labels and title
plt.xlabel('Value')
plt.ylabel('Count')
plt.title('Histogram with element="step"')

# Show plot
plt.show()
```
