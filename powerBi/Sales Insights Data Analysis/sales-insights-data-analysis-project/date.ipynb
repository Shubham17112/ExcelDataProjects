```python
import pandas as pd
from sqlalchemy import create_engine, text

# Quick check
def query_execute(query, return_result=True):
    engine = create_engine("mysql+pymysql://root:1234@localhost/sales")
    with engine.connect() as connection:
        if return_result:
            return pd.read_sql(query, connection)
        else:
            connection.execute(text(query))
            connection.commit()
            
```

```python
# anaylsing transctions table  ( data type is currect of each column)
describe_date = query_execute("select * from date")
describe_date
```

```python
# anaylsing transctions table  ( data type is currect of each column)
describe_date = query_execute("describe date")
describe_date
```

```python
# anaylsing transctions table  ( data type is currect of each column)
describe_date = query_execute("select count(*) from date")
describe_date
```
