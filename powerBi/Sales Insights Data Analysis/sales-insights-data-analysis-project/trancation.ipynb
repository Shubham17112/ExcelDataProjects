```python
import pandas as pd
from sqlalchemy import create_engine, text

# Quick check
def query_execute(query, return_result=True):
    engine = create_engine("mysql+pymysql://root:1234@localhost/sales")
    with engine.connect() as connection:
        if return_result:
            return pd.read_sql(query, connection)
        else:
            connection.execute(text(query))
            connection.commit()
            
```

```python
# anaylsing transctions table  ( data type is currect of each column)
describe_trancation = query_execute("describe transactions")
describe_trancation.print_()
```

```python
# there is no null values in the table
describe_trancation = query_execute("""SELECT * FROM transactions
WHERE product_code IS NULL
   OR customer_code IS NULL
   OR market_code IS NULL
   OR order_date IS NULL
   OR sales_qty IS NULL
   OR sales_amount IS NULL
   OR currency IS NULL;
""") 
describe_trancation.print_()
```
